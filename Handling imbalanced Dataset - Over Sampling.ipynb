- OverSampling is nothing but multiplying your test data for ex: if your train data is of 90% and test data is of 10% and you can say that the ratio is 9:1 then it is considered as an imbalanced data.
- OverSampling in this case is it will multiply each test data points 9 times and make the ratio 1:1
- Usually Over Sampling is used than Under Sampling
